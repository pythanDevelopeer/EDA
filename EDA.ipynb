```python
diamonds=pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

# plotting the histogram based price

```python
sns.distplot(diamonds['price'])
plt.show();
```

# Printing only the frequency cure

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

# adding color to the distribution plot

```python
sns.distplot(diamonds['price'],color='green')
plt.show()
```

# plotting only the histogram without the frequency curve

```python
sns.distplot(diamonds['price'],kde=False,color='green')
plt.show()
```

# changing the number of bins

```python
sns.distplot(diamonds['price'],kde=False,color='green',bins=10)
plt.show()
```

```python
sns.distplot(diamonds['price'],kde=False,color='green',bins=10,vertical=True)
plt.show()
```

# Boxplot

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

# Ploting based on the churn and tenure columns

```python
sns.boxplot(x='Churn',y='tenure',data=df)
plt.show()
```

# Plotting based on the internet and service 

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=df)
plt.show()
```

# plotting on the contract

```python
sns.boxplot(x='Contract',y='tenure',data=df,palette='Set1')
plt.show()
```

# changing line width of boxplot

```python
sns.boxplot(x='Contract',y='tenure',data=df,palette='Set1',linewidth=5)
plt.show()
```

# plotting based on custom ordering on x axis

```python
sns.boxplot(x='Contract',y='tenure',data=df,palette='Set1',linewidth=5,order=['One year','Two year'],hue='PaymentMethod')
plt.show()
```
